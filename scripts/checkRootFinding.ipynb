```python
import numpy as np
```

```python
%%timeit
pol_roots=[]
for pol in poly_list:
    f = simplify(pol)
    allroots = solveset(f)
    la =  Intersection(allroots,ConditionSet(x, np.abs(x)<=(1/np.sqrt(2)) , S.Complexes ))
    try:
        lam = la.args[0]
        pol_roots.append(lam)
    except:
        pol_roots.append(0)
print(len(pol_roots))
```

```python
%%timeit
pol_roots_g = []
for pol in poly_list:
    g = poly(pol)
    allroots = solve(g)
    for lamb in allroots:
        if Abs(lamb.evalf(30))<=(1/sqrt(2)).evalf(30):
            pol_roots_g.append(lamb)
print(len(pol_roots_g))
```
