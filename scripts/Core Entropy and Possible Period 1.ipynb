```python
import math
import matplotlib.pyplot as plt
from src.functions import core_entropy

def farey_sequence(n: int) -> list:
    """Return the nth Farey sequence."""
    (a, b, c, d) = (0, 1, 1, n)
    seq = []
    while (c <= n):
        k = (n + b) // d
        (a, b, c, d) = (c, d, k * c - a, k * d - b)
        if 2*a<b: seq.append([a, b])
    return seq
```

```python
angles = farey_sequence(75)
X = [key[0]/key[1] for key in angles]
Y = [core_entropy(num=key[0],den=key[1]) for key in angles]
```

```python
plt.scatter(X,Y, color='blue', alpha=0.5,marker=".")
plt.grid(True)
plt.title('Core Entropy for angles up to 1/2')
# plt.xlim(0.4,0.5)
# plt.ylim(1.4, 2)
plt.show()
```

```python
angles_1 = [] # [Angle(th=ang) for ang in data["1"]]
for ang in data["1"]:
    theta = Angle(th=ang)
    
    theta.per_len = data["1"][ang]["ks_per"]
    theta.start_index_per = data["1"][ang]["start_index_per"]
#     theta.ks = data["1"][ang]["ks"]
#     theta.itin = data["1"][ang]["itin"]
#     theta.itin_per_len = data["1"][ang]["itin_per"]
    
#     theta.rat_func = data["1"][ang]["rat_func"]
#     theta.rat_func = theta.itin_to_rat() # make sure to use ** instead of ^
    
    angles_1.append(theta)
```

```python
angles_1_lambda = []
angles_1_core = []

for ang in angles_1:
    lam = complex(ang.assoc_lambda())
    if abs(lam)>0.495:
        lam = lam if lam.imag>0 else lam.conjugate() 
        angles_1_lambda.append(ang.num/ang.den)
        angles_1_core.append(core_entropy(angle=ang))
```

```python
plt.scatter(X,Y,color="blue",alpha=0.5,marker='+')
plt.scatter(angles_1_lambda,angles_1_core,color='red',alpha=0.45, marker='x')
plt.grid(True)
plt.title('Core Entropy for angles up to 1/2 and possible associated parameters with period 1')

# plt.xlim(0.2,0.3)
# plt.ylim(1.6, 1.7)
plt.show()
```
